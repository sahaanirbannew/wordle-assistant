```python
from google.colab import drive
import os
import csv
import time
import pickle
drive.mount("/content/drive", force_remount=False)
ext_path = "/content/drive/My Drive/WordleProject/extracted/"
ext_path_process_log = "/content/drive/My Drive/WordleProject/process_log/"
filename_done_word_analytics = 'files_done_word_analytics.txt'
ext_path_word_analytics = "/content/drive/My Drive/WordleProject/word_analytics/"


def fetch_filenames():
  files = os.listdir(ext_path_process_log)
  files = set(files)
  return files

def get_individual_file_stats(filename):
  individual_dict={}
  with open(ext_path_process_log+filename, mode='r') as infile:
      reader = csv.reader(infile)
      for rows in reader:
          k = rows[0]
          v = rows[1]
          individual_dict[k]=v 
  
  return individual_dict

def fetch_main_dict():
  try:
    ext_path_word_analytics = "/content/drive/My Drive/WordleProject/word_analytics/"
    main_dict_filename = "main_dict.pkl"
    file = open(ext_path_word_analytics+main_dict_filename, 'rb')
    data = pickle.load(file)
    file.close()
    return data
  except: 
    print("Info: Waiting for the file to appear.")
    time.sleep(30)
    fetch_main_dict()


def add_dict_to_main(main_dict, individual_dict1): 
  for elem in individual_dict1:
    try: 
      if elem not in main_dict: main_dict[elem] = int(individual_dict1[elem])
      else: main_dict[elem] = int(main_dict[elem]) +int(individual_dict1[elem])
    except Exception as e :
      add_to_error_log("Error:"+elem + " " + individual_dict1[elem] + " " + str(e))
      print(e)
  return main_dict

def save_main_dict(main_dict):
  file = open(ext_path_word_analytics+'main_dict.pkl', 'wb')
  pickle.dump(main_dict, file)
  file.close()

def save_done_files(filenama):
  with open((ext_path_word_analytics+filename_done_word_analytics), "a") as writefile:
    writefile.write(filenama+",")
    writefile.close()

def add_to_error_log(text):
  with open((ext_path_word_analytics+"error_log.txt"), "a") as writefile:
    writefile.write(text+"\n")
    writefile.close()

def read_done_processing_1_file():
  try:
    with open((ext_path_word_analytics+filename_done_word_analytics), "r") as readfile:
      list_of_processed_files_array= readfile.read().split(",")
      while '' in list_of_processed_files_array:
        list_of_processed_files_array.remove('')
      return set(list_of_processed_files_array)
  except:
    print("Info: Waiting for the file to appear.")
    time.sleep(30) #in case the file is not saved so quickly on Google Drive.
    read_done_processing_1_file()
```

```python
#initialising
main_dict={'a':1}
save_main_dict(main_dict)
save_done_files("")

while 1>0:
  total_files = fetch_filenames() #fetches all file names 
  list_of_processed_files_array = read_done_processing_1_file() #fetches all processed file names
  files_to_process = total_files - list_of_processed_files_array #finds new files to process
  main_dict=fetch_main_dict() #loads main dict
  count= len(list_of_processed_files_array)
  for file in files_to_process: 
    individual_dict1=get_individual_file_stats(file) #fetches individual file statistics
    add_dict_to_main(main_dict, individual_dict1) #adds to the main dictionary
    save_done_files(file) #saves the processed filename to a file
    count +=1
    print(count, file)
    if (count%100 == 0): save_main_dict(main_dict)
  save_main_dict(main_dict) #after the loop saves the main dict
  print("sleeping for 60 seconds")
  time.sleep(60)
```
