```python
from google.colab import drive
import os
drive.mount("/content/drive", force_remount=False)
ext_path = "/content/drive/My Drive/WordleProject/extracted/"
ext_path_process_log = "/content/drive/My Drive/WordleProject/process_log/"
filename_done_processing_1 = 'files_done_processing_1.txt'
```

```python
import re
```

```python
def fetch_filenames():
  files = os.listdir(ext_path)
  files = set(files)
  return files

def save_done_files(title):
  with open((ext_path_process_log+filename_done_processing_1), "a") as writefile:
    writefile.write(title+",")
    writefile.close()

def read_done_processing_1_file():
  with open((ext_path_process_log+filename_done_processing_1), "r") as readfile:
    list_of_processed_files_array= readfile.read().split(",")
    while '' in list_of_processed_files_array:
      list_of_processed_files_array.remove('')
    return set(list_of_processed_files_array)
```

```python
def make_dict(filename):
  five_lettered_word = {}
  with open((ext_path+filename), "r") as readfile:
    content=readfile.read().replace("\n"," ").lower()
    content = re.sub('[^0-9a-z]+', ' ', content).split(" ")
    for word in content:
      if len(word)==5:
        if word not in five_lettered_word:
          five_lettered_word[word]= 1 
        else:
          five_lettered_word[word]= five_lettered_word[word] + 1 
    return five_lettered_word
```

```python
def export_dict(filename, dict_five_lettered_words):
  filename = filename.replace(".txt",".csv")
  with open((ext_path_process_log+filename), "w") as writefile:
    #Soham, this should be: from dictionary to CSV. (so that you can import csv as pandas dataframe.)

    with open(ext_path_process_log+filename, 'w') as f:
        for key in dict_five_lettered_words.keys():
            f.write("%s, %s\n" % (key, dict_five_lettered_words[key]))

    f.close()
```

```python
import time
while (1>0):
  total_files = fetch_filenames()
  list_of_processed_files_array = read_done_processing_1_file()
  files_to_process = total_files - list_of_processed_files_array

  if len(files_to_process)>0:
    print(len(total_files))
    counter = len(list_of_processed_files_array)
    print(len(list_of_processed_files_array))
    print(len(files_to_process))

    for filename in files_to_process:
      dict_five_lettered_words= make_dict(filename)  
      export_dict(filename,dict_five_lettered_words)
      save_done_files(filename)
      counter +=1
      print(counter)

  else:
    print("Going to sleep for 1 minute..")
    time.sleep(60)
      
```
